```python
#Introduction to Machine Learning-1 Assignment
"""Explain the following with an example:
1) Artificial Intelligence
2) Machine Learning,
3) Deep Learning"""

Ans: 
1) Artificial Intelligence (AI): Artificial Intelligence is a branch of computer science that focuses on 
creating intelligent machines that can think and act like humans. AI systems are designed to learn from 
their environment and adapt to new situations. An example of AI is a self-driving car, which uses sensors 
and algorithms to navigate its environment and make decisions about how to drive.

2) Machine Learning: Machine Learning is a subset of AI that focuses on developing algorithms that can 
learn from data and make predictions or decisions without being explicitly programmed. An example of 
machine learning is a facial recognition system, which uses algorithms to identify faces in images or 
videos.

3) Deep Learning: Deep Learning is a subset of machine learning that uses artificial neural networks to 
learn from data and make decisions. An example of deep learning is a computer vision system, which uses 
neural networks to identify objects in images or videos.

"""Q2-What is supervised learning? List some example of supervised learning."""
Ans: Supervised learning is a type of machine learning algorithm that uses labeled data to train a model 
to make predictions. Supervised learning algorithms learn from labeled data, which means that the data 
has been labeled with the correct output. Examples of supervised learning include linear regression, 
logistic regression, decision trees, support vector machines, and neural networks.

Examples of supervised learning include: 
- Linear regression 
- Logistic regression 
- Decision trees 
- Support vector machines 
- Neural networks 
- Naive Bayes classifier 
- K Nearest Neighbors (KNN) 
- Random forest 
- Gradient boosting

"""Q3 What is unsupervised learning? List some example of unsupervised learning."""
Ans: Unsupervised learning is a type of machine learning algorithm used to draw inferences from datasets 
consisting of input data without labeled responses. Examples of unsupervised learning include clustering, 
anomaly detection, and association rule learning.

#Examples of unsupervised learning include:

1. Clustering: Grouping data points into clusters based on similarity.
2. Anomaly Detection: Identifying outliers or anomalies in a dataset.
3. Association Rule Learning: Finding relationships between variables in a dataset.
4. Dimensionality Reduction: Reducing the number of features in a dataset.
5. Autoencoders: Compressing data into a lower-dimensional representation.
6. Generative Adversarial Networks (GANs): Generating new data from existing data.

"""Q4 What is the difference between AI, ML, DL, and DS?"""
Ans: 
AI (Artificial Intelligence): AI is a broad term that refers to the ability of machines to perform tasks 
that normally require human intelligence. AI can be used to solve complex problems, automate processes, 
and make decisions.

ML (Machine Learning): ML is a subset of AI that focuses on the development of algorithms that can learn 
from data and improve over time. ML algorithms are used to identify patterns in data and make predictions 
or decisions.

DL (Deep Learning): DL is a subset of ML that uses neural networks to process large amounts of data and 
make predictions or decisions. DL algorithms are used to identify complex patterns in data and can be used
for tasks such as image recognition, natural language processing, and autonomous driving.

DS (Data Science): DS is a field of study that focuses on the analysis of data to gain insights and make 
decisions. DS involves the use of various techniques such as data mining, machine learning, and statistical
analysis to extract meaningful information from data.

"""Q5- What are the main differences between supervised, unsupervised, and semi-supervised learning?"""
Ans: Supervised learning is a type of machine learning algorithm that uses labeled data to make predictions
. It relies on a set of labeled training data to learn how to map input data to output labels. Supervised 
learning algorithms can be used for classification and regression tasks.

Unsupervised learning is a type of machine learning algorithm that uses unlabeled data to make predictions.
It relies on a set of unlabeled training data to learn how to map input data to output labels. 
Unsupervised learning algorithms can be used for clustering and dimensionality reduction tasks.

Semi-supervised learning is a type of machine learning algorithm that uses both labeled and unlabeled data
to make predictions. It relies on a set of labeled and unlabeled training data to learn how to map input 
data to output labels. Semi-supervised learning algorithms can be used for classification, regression, 
clustering, and dimensionality reduction tasks.

"""Q6 What is train, test and validation split? Explain the importance of each term."""
Ans: Train, test, and validation split is a method of splitting a dataset into three distinct parts. The 
train set is used to train the model, the test set is used to evaluate the model's performance, and the 
validation set is used to tune the model's hyperparameters. This split is important because it allows the 
model to be tested on unseen data, which helps to prevent overfitting. Additionally, the validation set 
helps to ensure that the model is not over-optimized for the training data.

"""Q7 How can unsupervised learning be used in anomaly detection?"""
Ans: 
Unsupervised learning can be used in anomaly detection by clustering data points and then identifying 
outliers that do not fit into the clusters. This can be done by using algorithms such as k-means 
clustering, hierarchical clustering, or density-based clustering. These algorithms can be used to 
identify data points that are significantly different from the rest of the data, which can indicate 
anomalies.

Another way unsupervised learning can be used in anomaly detection is by using autoencoders. Autoencoders
are neural networks that are trained to reconstruct input data. Anomalies can be detected by comparing 
the reconstructed data with the original input data and identifying any discrepancies.

"""Q8 List down some commonly used supervised learning algorithms and unsupervised learning algorithms."""
Ans: 
#Supervised Learning Algorithms: 
1. Linear Regression 
2. Logistic Regression 
3. Decision Trees 
4. Support Vector Machines 
5. Naive Bayes Classifier 
6. K-Nearest Neighbors (KNN) 
7. Random Forest 

#Unsupervised Learning Algorithms: 
1. K-Means Clustering 
2. Hierarchical Clustering 
3. Principal Component Analysis (PCA) 
4. Singular Value Decomposition (SVD) 
5. Association Rule Learning 
6. Autoencoders 
7. Reinforcement Learning
```
